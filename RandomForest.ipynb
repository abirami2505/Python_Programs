```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('/content/Book1.csv')
```

```python
features = df[['Literacy','Cleanliess','Crime_Rate']]
Target = df.Output
```

```python
RF = RandomForestClassifier()
RF.fit(features,Target)
```

```python
l=int(input("Enter Literacy:"))
c1=int(input("Enter Cleanliess:"))
cr=int(input("Enter Crime_Rate: "))
pred= RF.predict([[l,c1,cr]])
if(pred==1):
  print("Output: Good")
else:
  print("Output: Bad")
```

```python
estimator = RF.estimators_[0]
```

```python
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=[20,20])
plot_tree(estimator,filled=True)
```

```python
fig = plt.figure(figsize=[20,10])
for i in range(5):
  for j in range(5):
    fig.add_subplot(5,2,1*2+j+1)
    plot_tree(RF.estimators_[i*2+j],feature_names=['Literacy','Cleanliess','Crime_Rate'],filled=True)
```

```python
from matplotlib import pyplot as plt
for i in range(0,10):
  plot_tree(RF.estimators_[i],feature_names=['Literacy','Cleanliess','Crime_Rate'],filled=True)
  plt.show()
```
