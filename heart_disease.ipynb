```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
hrt = pd.read_csv('/content/heart_failure_clinical_records_dataset.csv')
hrt.head()
```

```python
hrt.isnull().sum()
```

```python
ind = hrt[[	'age','diabetes','sex','smoking']]
dep = hrt['DEATH_EVENT']
```

```python
LR = LinearRegression()
```

```python
LR.fit(ind,dep)
```

```python
age = int(input("Enter the age: "))
diabetes = int(input("Enter the diabets:"))
sex = int(input("Enter the sex: "))
smoking = int(input("Enter the smoking:"))
pred = LR.predict([[age,diabetes,sex,smoking]])
print(pred)
```
