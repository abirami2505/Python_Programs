```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
df=pd.read_csv('/content/Book1.csv')
print(df.head)
```

```python
df.isnull().sum()
```

```python
feature = df[['Literacy','Cleanliess','Crime_Rate']]
Target = df['Output']
```

```python
DT=DecisionTreeClassifier(criterion='entropy')
DT.fit(feature,Target)
```

```python
l=int(input('Enter Literacy Rate: '))
c1 = int(input('Enter Cleanliess Rate: '))
cr = int(input('Enter Crime Rate: '))
pred = DT.predict([[l,c1,cr]])
print(pred)
```

```python
plot_tree(DT)
```
