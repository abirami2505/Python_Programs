```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
dia = pd.read_excel('/content/diabetes (1).xlsx')
dia.head()
```

```python
dia.isnull().sum()
```

```python
ind = dia[[	'age','mass','insu','skin']]
dep = dia['class']
```

```python
Logr = LogisticRegression()
```

```python
Logr.fit(ind,dep)
```

```python
age = int(input("Enter the age: "))
mass = int(input("Enter the mass:"))
insu = int(input("Enter the insu: "))
skin = int(input("Enter the skin:"))
pred = Logr.predict([[age,mass,insu,skin]])
print(pred)
```
